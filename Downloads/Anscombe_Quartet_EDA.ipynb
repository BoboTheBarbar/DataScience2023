## Below is a helper function that takes the anscombe quartet dataset and returns 4 dataframes. 1 for each of the ancombe's set

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


def separate_datasets(anscombe):
    """
    This function takes the Anscombe's quartet dataset loaded using Seaborn and returns four separate Pandas dataframes, 
    one for each of the four datasets.
    
    Args:
    - anscombe (Pandas dataframe): The Anscombe's quartet dataset loaded using Seaborn
    
    Returns:
    - dataset_I, dataset_II, dataset_III, dataset_IV (Pandas dataframes): Four separate dataframes, one for each of the four datasets.
    """
    
    dataset_I = anscombe[anscombe['dataset'] == 'I'].reset_index(drop=True).drop('dataset', axis=1)
    dataset_II = anscombe[anscombe['dataset'] == 'II'].reset_index(drop=True).drop('dataset', axis=1)
    dataset_III = anscombe[anscombe['dataset'] == 'III'].reset_index(drop=True).drop('dataset', axis=1)
    dataset_IV = anscombe[anscombe['dataset'] == 'IV'].reset_index(drop=True).drop('dataset', axis=1)
    
    return dataset_I, dataset_II, dataset_III, dataset_IV
```
